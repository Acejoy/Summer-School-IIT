```python
import datetime
import cv2
import time
import imutils
import numpy as np
import matplotlib.pyplot as plt
from imutils.video import VideoStream
```

```python
video_path=r'C:\Users\Legion\Desktop\Internship_IIT\Crowd-Activity-All.avi'
min_area=500
vs=cv2.VideoCapture(video_path)
firstFrame=None
prev_gray=None
while True:

    frame=vs.read()
    if frame is None:
        break
    else:
        frame=frame[1]
        
    frame=imutils.resize(frame,width=500)
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray=cv2.GaussianBlur(gray,(21,21),0)
    
    if prev_gray is None:
        prev_gray = gray
        continue

    #print('hey1')
    frameDelta = cv2.absdiff(prev_gray,gray)
    thresh=cv2.threshold(frameDelta,25,255,cv2.THRESH_BINARY)[1]
    
    thresh=cv2.dilate(thresh,None,iterations=2)
    cnts=cv2.findContours(thresh.copy(),cv2.RETR_EXTERNAL,
                         cv2.CHAIN_APPROX_SIMPLE)
    cnts=imutils.grab_contours(cnts)
    #print('hey2')
    for c in cnts:
        
        if cv2.contourArea(c) < min_area:
            continue
        
        #print('hey3')
        (x,y,w,h)=cv2.boundingRect(c)
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
        
    cv2.imshow("frames",frame)
    cv2.imshow('thrsh',thresh)
    cv2.imshow('frame delta',frameDelta)
    print('hey4')
    key=cv2.waitKey(5) & 0xFF
    #print('hey5')
    prev_gray=gray
    #time.sleep(0.01)
    if key == ord("q"):
        break
        
vs.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
cap = cv.VideoCapture(video_path)
fgbg = cv.bgsegm.createBackgroundSubtractorMOG()
while(1):
    ret, frame = cap.read()
    fgmask = fgbg.apply(frame)
    l=cv2.bitwise_and(frame,frame,mask = fgmask)
    cv.imshow('frame',l)
    cv.imshow('orig frame',frame)
    k = cv.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
cap = cv.VideoCapture(video_path)
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE,(3,3))
fgbg = cv.bgsegm.createBackgroundSubtractorGMG()
while(1):
    ret, frame = cap.read()
    fgmask = fgbg.apply(frame)
    fgmask = cv.morphologyEx(fgmask, cv.MORPH_OPEN, kernel)
    l=cv2.bitwise_and(frame,frame,mask = fgmask)
    cv.imshow('frame',l)
    k = cv.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv.destroyAllWindows()
```
