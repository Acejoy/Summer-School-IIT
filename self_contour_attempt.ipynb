```python
import numpy as np
import cv2 
im_path1=r'C:\Users\Legion\Desktop\Internship_IIT\UCSD_Anomaly_Dataset.v1p2\UCSDped1\Train\Train001\023.tif'
im_path2=r'C:\Users\Legion\Desktop\Internship_IIT\PETS 2009\Dataset S0\S0_BG\Crowd_PETS09\S0\Background\View_001\Time_13-19\00000001.jpg'
im = cv2.imread(im_path1,1)
imgray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray
                           , 127, 255, cv2.THRESH_BINARY)
contours, heirar = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

rects = []

for cnt in contours:
    if cv2.contourArea(cnt)>20 and cv2.contourArea(cnt)<200:
        rects.append(cv2.boundingRect(cnt))

for rect in rects:
    x,y,w,h = rect
    cv2.rectangle(im, (x,y), (x+w,y+h),(0,255,0),1)
cv2.imshow('show',im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
