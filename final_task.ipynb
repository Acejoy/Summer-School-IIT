```python
import os
import cv2
import time
import numpy as np
from sklearn.decomposition import PCA
from numpy.linalg import norm, svd
```

```python
def extract_dominant_dir(dir_matrix):
    pca = PCA()
    pca.fit(dir_matrix)
    components = pca.components_
        
    return np.mean(components[1,:])       
```

```python
M = 20
N = 20

def extract_feature_vecs(flow):
    
    x,y = flow[...,0],flow[...,1]
    mag, direction = cv2.cartToPolar(x, y)
    
    M_step = 20
    N_step = 20
    
    lower_x = 0
    lower_y = 0
    
    r1 = []
    r2 = []
    index = 0
    while(lower_x < flow.shape[0]):
        
        while( lower_y < flow.shape[1]):
            #print('x_lower:',lower_x," y_lower:",lower_y)
            temp_mag_matrix = mag[lower_x: lower_x+ M_step, lower_y: lower_y+ N_step]
            temp_dir_matrix = direction[lower_x: lower_x+ M_step, lower_y: lower_y+ N_step]
            print('index is :',index,'from (',lower_x,lower_y,') to (',lower_x+M_step,lower_y+N_step,')')
            index+=1
            blockwise_mean = np.mean(temp_mag_matrix)
            r1.append(blockwise_mean)        
            r2.append(extract_dominant_dir(temp_dir_matrix))
            
            
            lower_y += N_step
                
        lower_y = 0
        lower_x += M_step
        
        
    r1 = np.asarray(r1)
    r2 = np.asarray(r2)
    
    return r1,r2    
    
    
```

```python
#calculating the dense optical flow using farneback algorithm
frameset_path = input("Enter the frame directory:")

list_frames_name = os.listdir(frameset_path)

if '._.DS_Store' in list_frames_name:
    list_frames_name.remove('._.DS_Store')
    
if '.DS_Store' in list_frames_name:
    list_frames_name.remove('.DS_Store')
    

    
ref_frame_path = os.path.join(frameset_path,list_frames_name[0])
ref_frame = cv2.imread(ref_frame_path,0)
#ref_frame = cv2.resize(ref_frame,(240,240))

feature_matrix_1 = []
feature_matrix_2 = []


for frame_num, frame_name in enumerate(list_frames_name[1:]):
    frame_path = os.path.join(frameset_path,frame_name)
    print(frame_num,'------>')
    frame = cv2.imread(frame_path, 0)
    #frame = cv2.resize(frame,(240,240))
    flow = cv2.calcOpticalFlowFarneback(ref_frame, frame, None, 0.5, 5, 15, 3, 5, 1.1, cv2.OPTFLOW_FARNEBACK_GAUSSIAN)
    
    r1, r2 = extract_feature_vecs(flow)
    
    feature_matrix_1.append(r1)
    feature_matrix_2.append(r2)
    
    
#transpose them

feature_matrix_1 = np.asarray(feature_matrix_1)
feature_matrix_2 = np.asarray(feature_matrix_2)

feature_matrix_1 = feature_matrix_1.T
feature_matrix_2 = feature_matrix_2.T



```

```python
#defining the inexact alm algorithm

def inexact_augmented_lagrange_multiplier(X, lmbda=.01, tol=1e-3,
                                          maxiter=100, verbose=True):
    """
    Inexact Augmented Lagrange Multiplier
    """
    lmbda = 1/np.sqrt(X.shape[0])
    
    Y = X
    norm_two = norm(Y.ravel(), 2)
    norm_inf = norm(Y.ravel(), np.inf) / lmbda
    dual_norm = np.max([norm_two, norm_inf])
    Y = Y / dual_norm
    A = np.zeros(Y.shape)
    E = np.zeros(Y.shape)
    dnorm = norm(X, 'fro')
    mu = 1.25 / norm_two
    rho = 1.5
    sv = 10.
    n = Y.shape[0]
    itr = 0
    while True:
        Eraw = X - A + (1 / mu) * Y
        Eupdate = np.maximum(Eraw - lmbda / mu, 0) + np.minimum(Eraw + lmbda / mu, 0)
        U, S, V = svd(X - Eupdate + (1 / mu) * Y, full_matrices=False)
        svp = (S > 1 / mu).shape[0]
        if svp < sv:
            sv = np.min([svp + 1, n])
        else:
            sv = np.min([svp + round(.05 * n), n])
        Aupdate = np.dot(np.dot(U[:, :svp], np.diag(S[:svp] - 1 / mu)), V[:svp, :])
        A = Aupdate
        E = Eupdate
        Z = X - A - E
        Y = Y + mu * Z
        mu = np.min([mu * rho, mu * 1e7])
        itr += 1
        if ((norm(Z, 'fro') / dnorm) < tol) or (itr >= maxiter):
            break
    if verbose:
        print("Finished at iteration %d" % (itr))  
    return A, E
```

```python
#GODEC algorithm

import numpy as np
from numpy.linalg import norm
from scipy.linalg import qr

def wthresh(a, thresh):
    #Soft wavelet threshold
    res = np.abs(a) - thresh
    return np.sign(a) * ((res > 0) * res)

#Default threshold of .03 is assumed to be for input in the range 0-1...
#original matlab had 8 out of 255, which is about .03 scaled to 0-1 range
def go_dec(X, thresh=.03, rank=2, power=0, tol=1e-3,
           max_iter=100, random_seed=0, verbose=True):
    m, n = X.shape
    if m < n:
        X = X.T
    m, n = X.shape
    L = X
    S = np.zeros(L.shape)
    itr = 0
    random_state = np.random.RandomState(random_seed)    
    while True:
        Y2 = random_state.randn(n, rank)
        for i in range(power + 1):
            Y1 = np.dot(L, Y2)
            Y2 = np.dot(L.T, Y1);
        Q, R = qr(Y2, mode='economic')
        L_new = np.dot(np.dot(L, Q), Q.T)
        T = L - L_new + S
        L = L_new
        S = wthresh(T, thresh)
        T -= S
        err = norm(T.ravel(), 2)
        if (err < tol) or (itr >= max_iter):
            break 
        L += T
        itr += 1
    #Is this even useful in soft GoDec? May be a display issue...
    G = X - L - S
    if m < n:
        L = L.T
        S = S.T
        G = G.T
    if verbose:
        print("Finished at iteration %d" % (itr))
    return L, S, G
```
